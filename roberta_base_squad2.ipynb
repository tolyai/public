```python
def question(filename):
  print(filename)
  with open(f'/content/drive/My Drive/LWProject/contracts_txt/{filename}', 'r') as file:
    data_x = file.read()
  QA_input = {
    'question': 'When is expiration date?',
    'context': data_x
    }
  res = nlp(QA_input)
  return res
```

```python
results = {'answer':[],
           'score':[],
           'filename':[],
           'start':[],
           'end':[]}

for fl in files:
    with open(f'/content/drive/My Drive/LWProject/contracts_txt/{fl}', 'r') as file:
      data_x = file.read()
    QA_input = {
    'question': 'When is expiration date?',
    'context': data_x
    }
    res = nlp(QA_input)
    results['answer'].append(res['answer'])
    results['score'].append(res['score'])
    results['filename'].append(fl)
    results['start'].append(res['start'])
    results['end'].append(res['end'])
```

```python
results = {'answer':[],
           'score':[],
           'filename':[],
           'start':[],
           'end':[]}

for fl in onlyfiles:
    with open(f'/content/drive/My Drive/LWProject/contracts_txt/{fl}', 'r') as file:
      data_x = file.read()
    QA_input = {
    'question': 'When is expiration date?',
    'context': data_x
    }
    res = nlp(QA_input)
    results['answer'].append(res['answer'])
    results['score'].append(res['score'])
    results['filename'].append(fl)
    results['start'].append(res['start'])
    results['end'].append(res['end'])
```
